```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 5)

pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
```

```python
url='https://drive.google.com/file/d/1ggRw-tVMCSCiqz5sw8If5f3vhPlHxW2f/view?usp=sharing'
```

```python
RawData = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
```

```python
complaints = pd.read_csv(RawData)
```

```python
complaints.head(25)
```

```python
noise_complaints = complaints[complaints['Descriptor'] == "Loud Talking"]
noise_complaints[:10]
```

```python
noise_complaints['Borough'].value_counts().plot.bar(x='Cities', y='Loud Talking Compliants', rot=1, color='b')
```

```python
noise_complaints['Borough'].value_counts().plot(kind='barh')
```

```python
noise_complaints['Borough'].value_counts().plot(kind='pie')
```
